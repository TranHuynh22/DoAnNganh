# **Tiền Xử Lý Dữ Liệu**

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

file = "/content/drive/MyDrive/Colab Notebooks/DoAnNganh/df_Titanic.xlsx"
df_Titanic = pd.read_excel(file)

df_Titanic.info()
df_Titanic
```

```python
# Kiểm tra dữ liệu trùng lặp
duplicates = df_Titanic[df_Titanic.duplicated()]

if duplicates.empty:
    print("Không có dữ liệu trùng lặp.")
else:
    print("Có dữ liệu trùng lặp.")
    print(duplicates)
```

```python
# Kiểm tra dữ liệu trống
df_Titanic.isnull().sum()
```

```python
# Điền giá trị trung vị vào dữ liệu trống
df1_Titanic = df_Titanic.fillna(df_Titanic.select_dtypes(include=['number']).median(), inplace = True)
```

```python
# Kiểm tra dữ liệu trống
df_Titanic.isnull().sum()
```

```python
df_Titanic.drop(columns=['Cabin'], inplace=True)

# Kiểm tra dữ liệu trống
df_Titanic.isnull().sum()
```

```python
# Mã hóa dữ liệu
df_Titanic = pd.get_dummies(df_Titanic, columns=['Sex', 'Embarked'], drop_first=True)
```

```python
df_Titanic.drop(columns=['Name', 'Ticket', 'PassengerId'], inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler

# Chuẩn hóa dữ liệu
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_Titanic)
```

# **Mối Quan Hệ Tương Quan**

```python
f, ax = plt.subplots(figsize=(15,6))
sns.heatmap(df_Titanic.corr(), annot=True, ax=ax, lw=.7, linecolor="black", fmt=".2f")
plt.show()
```

# **Trực Quan Hóa Dữ Liệu Trước Khi Thực Hiện PCA**

```python
plt.scatter(df_Titanic.iloc[:, 0], df_Titanic.iloc[:, 1], c=df['Survived'], cmap='coolwarm')
plt.xlabel('Sex')
plt.ylabel('Fare')
plt.title('Trực quan hóa dữ liệu Titanic trước PCA')
plt.show()
```

# **Ứng dụng PCA**

```python
from sklearn.decomposition import PCA

# Thực hiện PCA, giữ lại tất cả các thành phần
pca = PCA(n_components=None)
df_pca = pca.fit_transform(df_scaled)
```

```python
# Chuyển đổi thành DataFrame
#df_scaled = pd.DataFrame(df_scaled, columns=df_Titanic.columns)

# Xem ma trận tải (loading matrix)
loading_matrix = pca.components_

# Hiển thị các thuộc tính đóng góp vào PC
df_loading = pd.DataFrame(loading_matrix.T, columns=[f'PC{i+1}' for i in range(loading_matrix.shape[0])], index=df_Titanic.columns)
print(df_loading)
```

```python
import matplotlib.pyplot as plt

# Tỷ lệ phương sai được giải thích
explained_variance = pca.explained_variance_ratio_

# Vẽ biểu đồ
plt.plot(range(1, len(explained_variance) + 1), explained_variance.cumsum(), marker='o', linestyle='--')
plt.title('Tỷ lệ phương sai được giải thích')
plt.xlabel('Số lượng thành phần chính')
plt.ylabel('Tỷ lệ phương sai tích lũy')
plt.show()
```

```python
# So sánh phương sai giải thích
explained_variance = pca.explained_variance_ratio_
print("Phần trăm phương sai giải thích bởi mỗi thành phần chính:", explained_variance)

# Vẽ biểu đồ heatmap của ma trận tải
f, ax = plt.subplots(figsize=(15,6))
sns.heatmap(df_loading, annot=True, cmap='coolwarm')
plt.show()
```

```python
# Thực hiện PCA với số thành phần giữ lại 90% phương sai
pca = PCA(n_components=0.90)
df_pca_reduced = pca.fit_transform(df_scaled)

print(f'Số lượng thành phần chính được giữ lại: {df_pca_reduced.shape[1]}')
```

# **Trực Quan Hóa Dữ Liệu**

```python
plt.scatter(df_pca_reduced[:, 0], df_pca_reduced[:, 1], c=df['Survived'], cmap='coolwarm')
plt.xlabel('Thành phần chính 1')
plt.ylabel('Thành phần chính 2')
plt.title('Trực quan hóa dữ liệu Titanic sau PCA')
plt.show()
```

# **Huấn Luyện Mô Hình Học Máy**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Chia dữ liệu thành tập huấn luyện và kiểm tra
X_train, X_test, y_train, y_test = train_test_split(df_pca_reduced, df['Survived'], test_size=0.2, random_state=42)

# Huấn luyện mô hình Logistic Regression
model = LogisticRegression()
model.fit(X_train, y_train)

# Dự đoán và tính toán độ chính xác
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

print(f'Độ chính xác của mô hình sau khi áp dụng PCA: {accuracy}')
```
